```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/Salary_Data.csv')
df
```

```python
df.tail()
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color='r')
plt.xlabel('year of exp',color='g')
plt.ylabel('salray',color='b')
plt.title('yr of exp and salary',color='m')
plt.grid()
plt.legend()
```

```python
#regression doesnot require normalization
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
x_train
```

```python
y_train
```

```python
#model creation
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
x_test
```

```python
y_test
```

```python
plt.scatter(x_train,y_train,color='b')
plt.plot(x_test,y_pred,color='r')
```

```python
print(model.predict([[6.1]]))
```

```python
df1=pd.DataFrame({'Actual value':y_test,'predicted value':y_pred})
df1
```

```python
print('slope is',model.coef_)
```

```python
print('constant is',model.intercept_)
```

```python
#performance evaluation

#1.MAE (mean absolute error)
#2.MAPE(mean absolute precentage error)
#3.MSE(mean squared error)
#4.RMSE(root mean squared error)
#5.R2_score(coefficient of determination )

```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print('error is',mean_absolute_percentage_error(y_test,y_pred))
print('error is',r2_score(y_test,y_pred))
```
